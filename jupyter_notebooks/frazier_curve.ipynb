```python
import matplotlib
import numpy
import scipy
from matplotlib import pyplot
from scipy import stats
```

```python
mean = -1
standard_deviation = 1.5

# Create Figure and Axes instances
fig,ax = pyplot.subplots(1)

x_values = numpy.arange(-5, 5, 0.1)
y_values = scipy.stats.norm(mean, standard_deviation)

ax.set_yticklabels([])
ax.set_xticklabels([])

ax.set_ylabel('How good you are')
ax.set_xlabel('How much of a dick you are')

frazier_curve = pyplot.plot(x_values, y_values.pdf(x_values))
frazier_curve
```
