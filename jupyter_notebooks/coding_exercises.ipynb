```python
members = [2, 7, 11, 15]
target = 18

def two_sum(list_of_members, target_value):
    for i in range(len(list_of_members)):
        if i >= len(list_of_members)-1:
            print("No more values to add.")
            return 0
        if list_of_members[i] + list_of_members[i+1] == target_value:
            print("Success...")
            return target
```

```python
DIVIDEND = 100
DIVISOR = -2

# There's a limitation to the size of the integer we are generating. This prevents us from dropping into an infinite loop.
MAX_ITERATIONS = 100000

def perform_division(dividend, divisor, max_iterations):
    NUM_LOOPS = 0
    result = 0
    
    # Dividing by zero isn't allowed. And, with our current logic, neither are negative dividends.     
    if divisor == 0:
        return "Invalid divisor passed. Cannot divide by zero"
    
    # Either both dividend and divisor are on the same side of zero, in which case the quotient is positive.
    # or there is a mismatch, and the quotient is negative.     
    if divisor > 0:
        modifier = 1
    else:
        print("Mismatch.")
        modifier = -1
    
    # A negative divisor breaks our division logic, so we make it positive. It's value will be reestablished later.     
    divisor *= modifier
    print(f"New divisor: {divisor}")
    
    for i in range(0, max_iterations):
        NUM_LOOPS += 1
        result += divisor
        if result >= dividend:
            if result > dividend:
                return (NUM_LOOPS - 1) * modifier
            return NUM_LOOPS * modifier
    
    return "There was a problem with the dividing logic, we have passed through the loop without generating a quotient."
```

```python
print(perform_division(DIVIDEND, DIVISOR, MAX_ITERATIONS))
```

```python
# Palidrome
TEN = 10
def reverse_integer(integer_to_flip):
    new_int = 0
    integer_to_flip_orig = integer_to_flip
    temp = 0
    
    # No point in attempting a palindrome on a single digit number.     
    if integer_to_flip < TEN:
        return integer_to_flip
    
    while integer_to_flip > 0:
        
        temp = integer_to_flip % TEN
        
        new_int = (new_int * TEN) + temp
        print(new_int)
        # Chip away at integer to flip         
        integer_to_flip //= TEN
    
    if new_int == integer_to_flip_orig:
        return True
    else:
        return False
    
```

```python
TEN = 10

class ListNode:
    def __init__(self, node_value=None):
        self.node_value = node_value
        self.next_value = None
        
class SingleLinkedList:
    def __init__(self):
        self.head_value = None
    
    def append_to_list(self, value):
        current = self.head_value
        if current:
            while current.next_value != None:
                current = current.next_value
            current.next_value = ListNode(value)
        else:
            self.head_value = ListNode(value)
    
    def print_list(self):
        val_to_print = self.head_value
        while val_to_print is not None:
            print(val_to_print.node_value)
            val_to_print = val_to_print.next_value


list_one = SingleLinkedList()
list_one.head_value = ListNode(2)
l1_n2 = ListNode(4)
l1_n3 = ListNode(3)

# Link to nodes:
list_one.head_value.next_value = l1_n2
l1_n2.next_value = l1_n3

list_two = SingleLinkedList()
list_two.head_value = ListNode(5)
l2_n2 = ListNode(6)
l2_n3 = ListNode(4)

# Link the nodes:
list_two.head_value.next_value = l2_n2
l2_n2.next_value = l2_n3

# This is defined in a cell above, but no reason to rewrite the code, or have to be dependent on that cell.
# This is jupyter land, after all. Plus, the return value for this method is different than its contemporary above.
def reverse_integer(integer_to_flip):
    new_int = 0
    integer_to_flip_orig = integer_to_flip
    temp = 0
    
    # No point in attempting a palindrome on a single digit number.     
    if integer_to_flip < TEN:
        return integer_to_flip
    
    while integer_to_flip > 0:
        
        temp = integer_to_flip % TEN
        
        new_int = (new_int * TEN) + temp

        # Chip away at integer to flip         
        integer_to_flip //= TEN
    
    return new_int

def render_list_to_int(linked_list):
    current_value = linked_list.head_value
    integer_to_return = 0
    while current_value is not None:
        integer_to_return = (integer_to_return * TEN) + current_value.node_value
        current_value = current_value.next_value
    return integer_to_return


def render_int_to_list(integer_to_listify):
    constituent_digits = []
    list_from_int = SingleLinkedList()
    while integer_to_listify > 0:
        temp = integer_to_listify % TEN
        list_from_int.append_to_list(temp)
        integer_to_listify //= TEN
    
    return list_from_int

list_one_integer = render_list_to_int(list_one)
list_one_flipped_int = reverse_integer(list_one_integer)

list_two_integer = render_list_to_int(list_two)
list_two_flipped_int = reverse_integer(list_two_integer)

desired_integer = list_one_flipped_int + list_two_flipped_int
test_list = render_int_to_list(desired_integer)
test_list.print_list()
```
